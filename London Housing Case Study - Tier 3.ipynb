**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
def price_increase_ratio(borough):
    average_price_2002 = properties_tall[properties_tall['Year']== 2002]
    average_price_2002_mean = pd.Series(average_price_2002[average_price_2002['boroughs'] == borough]
                                        .groupby('Year')['Average_prices'].mean()).values
    average_price_2022 = properties_tall[properties_tall['Year']== 2022]
    average_price_2022_mean = pd.Series(average_price_2022[average_price_2022['boroughs'] == borough]
                                        .groupby('Year')['Average_prices'].mean()).values
    increase_ratio = (average_price_2022_mean[0] / average_price_2002_mean[0])*100
    return increase_ratio

def average_price_ratio(borough):
    ### Function to determine the average price ratio between year 1998 and 2018 for user predefined borough ###
    
    average_price_1998 = properties_tall[properties_tall['Year']== 1998]
    average_price_1998_mean = pd.Series(average_price_1998[average_price_1998['boroughs'] == borough]
                                        .groupby('Year')['Average_prices'].mean()).values
    average_price_2018 = properties_tall[properties_tall['Year']== 2018]
    average_price_2018_mean = pd.Series(average_price_2018[average_price_2018['boroughs'] == borough]
                                        .groupby('Year')['Average_prices'].mean()).values
    
    ratio = average_price_1998_mean[0] / average_price_2018_mean[0]
    return ratio

london_boroughs = properties_tall.loc[properties_tall.Year == 2018,['boroughs','Average_prices']]
london_boroughs_unique = london_boroughs.drop_duplicates(['boroughs']).reset_index(drop=True).copy()
                                  
london_boroughs_unique['ratio_1998_2018'] = london_boroughs_unique.boroughs.map(average_price_ratio)
london_boroughs_unique['increase_ratio_2022_2002'] = london_boroughs_unique.boroughs.map(price_increase_ratio)                                            
print(london_boroughs_unique.sort_values(by='increase_ratio_2022_2002', ascending=False))
```

```python
#properties_tall.plot(kind='bar', x = 'boroughs', y = 'ratio_1998_2018')
london_boroughs_unique_st = london_boroughs_unique.sort_values('ratio_1998_2018', ascending = False)
london_boroughs_unique_st.plot(x = 'boroughs', y = 'ratio_1998_2018', kind='bar', figsize=(10,5))
```

```python
london_boroughs_unique_st = london_boroughs_unique.sort_values('increase_ratio_2022_2002', ascending = False)
london_boroughs_unique_st.plot(x = 'boroughs', y = 'increase_ratio_2022_2002', kind='bar',
                               ylabel ='percentage [%] of price increase', figsize=(10,5))
```
